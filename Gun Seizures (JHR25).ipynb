```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
guns = pd.read_csv('GunSeizures.csv')
gun=guns.dropna()
plt.figure(figsize=(20,10))
plt.bar(x=gun['neighborhood'],
        height=gun['total_count'])
plt.title('Total Number of Guns Seized by Police since 2015')
plt.xlabel('Neighborhoods')
plt.ylabel('Gun Seizures')
plt.xticks(rotation=90)
```

```python
gun.describe()
```

The data indicates that the minimum number of gun seizures in any neighborhood is 1. That means that neighboorhoods with 1 gun seizure are safest since gun violence is not prevelant. Either there are not many guns in these neighborhoods or at the very least there are not many guns being used to commit crimes. In either case, these neighborhoods are the safest to live in according to this metric. The safest neighborhoods with only 1 gun seizures can be seen on the bar graph and they are Westwood,Chartiers City,Polish Hill, Banksville,Stanton Heights,Regent Square, Oakwood, Mt Oliver, and West Ends.
